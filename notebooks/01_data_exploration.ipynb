## Step 3 — Understand the Structure

## Step 5 — Explore Key Columns

We focus on the most important columns for our analysis:
- Attrition (our target — did the employee leave?)
- Department
- JobRole
- OverTime

```python
# ── How many employees left vs stayed? ────────────────────────────
print("Attrition counts:")
print(df['Attrition'].value_counts())
print()
print("Attrition percentage:")
print(df['Attrition'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%')
```

```python
# ── What departments exist? ────────────────────────────────────────
print(f"Number of unique departments : {df['Department'].nunique()}")
print(df['Department'].value_counts())
print()

# ── What job roles exist? ──────────────────────────────────────────
print(f"Number of unique job roles   : {df['JobRole'].nunique()}")
print(df['JobRole'].value_counts())
```

```python
# ── Overtime distribution ──────────────────────────────────────────
print("Overtime:")
print(df['OverTime'].value_counts())
print()
print("Overtime percentage:")
print(df['OverTime'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%')
```
